```python
from matplotlib import pyplot as plt
import glob
import numpy as np
import os
import pandas as pd
import re
```

```python
md = pd.read_csv('metadata.csv', sep=';', encoding='utf-8')
```

```python
md
```

```python
files = []

for file in os.listdir('Texte'):
    files.append(file)
```

```python
data = []

for file in os.listdir('Texte'):
    with open(os.path.join('Texte', file),'r', encoding='utf-8', errors='ignore') as myfile:
        data.append(myfile.read())
```

```python
corpus = pd.DataFrame({'file': files, 'text': data})
```

```python
# Text von Tags befreien

def parser(text):
    return re.sub('<[^<]+?>', '', str(text))

corpus['text'] = corpus.text.apply(parser)
```

```python
df = pd.merge(md, corpus, how='left', left_on='doc', right_on='file')
```

```python
# Wörter zählen und hinzfügen

df['words'] = df['text'].str.count(r'\w+')
```

```python
df = df.fillna(0)
```

```python
# Spalten neu anordnen und unnötige entfernen

df = df[['doc', 'source', 'author', 'title', 'year', 'period', 'text type', 'text', 'words']]
```

```python
df.head()
```

```python
# Gutenberg Korpus

gb = pd.read_csv('Gutenberg.csv', sep=',', encoding='utf-8')
```

```python
gb.head()
```

```python
# Gutenberg Korpus anhängen

df_big = pd.concat([df, gb], axis=0, join='outer', ignore_index=False)
```

```python
df_big
```

```python
# chronologisch ordnen

df_big = df_big.sort_values(by=['year'])
```

```python
# Anzahl Texte pro Zeitraum

periods = list(df_big['period'].unique())
per_len ={i: len(df_big[df_big['period']==i]) for i in periods}
per_len
```

```python
plt.figure(figsize=[20,10])
plt.style.use('classic')
#plt.style.use('seaborn-whitegrid')

x_indexes = np.arange(len(periods))
plt.bar(x_indexes, per_len.values(), label='Zeiträume', width=0.8, color='b')
plt.xticks(ticks = x_indexes, labels = per_len.keys())
plt.xlabel('Zeiträume')
plt.ylabel('Anzahl der Texte')

plt.tight_layout()
plt.margins(0.2, 0.15)

for i, j in zip(x_indexes, per_len.values()):
       plt.annotate(j, xy=(i,j), xycoords='data', verticalalignment='bottom', horizontalalignment='center', 
                    fontweight='demi', fontsize='small')

plt.show()
```

```python
# Anzahl Wörter pro Zeitraum

sum_words = {i: sum(df_big[df_big['period']==i].words) for i in periods}
sum_words
```

```python
plt.figure(figsize=[20,10])
plt.style.use('classic')

x_indexes = np.arange(len(periods))
plt.bar(x_indexes, sum_words.values(), label='Zeiträume', width=0.8, color='b')
plt.xticks(ticks = x_indexes, labels = sum_words.keys())
plt.xlabel('Zeiträume')
plt.ylabel('Anzahl der Wörter')

plt.tight_layout()
plt.margins(0.2, 0.15)

for i, j in zip(x_indexes, sum_words.values()):
       plt.annotate(j, xy=(i,j), xycoords='data', verticalalignment='bottom', horizontalalignment='center', 
                    fontweight='demi', fontsize='small')

plt.show()
#plt.savefig('korpus.png', bbox_inches='tight')
```
