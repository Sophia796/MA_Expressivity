# Visualisierungen der manuellen Analyse

```python
from matplotlib import pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
```

```python
periods = ['1700-1750', '1751-1800', '1801-1850', '1851-1900', '1901-1950', '1951-2000', '2001-2021']
```

### terribilmente

```python
CU = [7, 6, 13, 48, 26, 2, 2]
BU = [0, 2, 10, 43, 61, 14, 23]
NU = [0, 0, 5, 17, 36, 98, 89]
PU = [0, 0, 0, 0, 2, 14, 11]
```

```python
plt.figure(figsize=[14,7])
plt.style.use('seaborn-whitegrid')

data = np.array([CU, BU, NU, PU])
labels = ['konventionell', 'ambig', 'intensivierend', 'positiv intensivierend']
colors = ['#00004d', '#0000FF', '#008fcc', '#80dfff']

X = np.arange(data.shape[1])
for i in range(data.shape[0]):
    plt.bar(X, data[i],bottom = np.sum(data[:i], axis=0),label=labels[i % len(labels)], color=colors[i % len(colors)])

x_indexes = np.arange(len(periods))
plt.xticks(ticks=x_indexes, labels=periods, rotation=45, horizontalalignment='right', fontsize='small')
#plt.xlabel('Zeitr√§ume')
#plt.ylabel('Use Case')
plt.ylim(0, 150)
plt.legend(loc='upper left', fontsize='large')

plt.tight_layout()
plt.margins(0.2, 0.15)

#plt.savefig('Plots/terribilmente_usecases.png')
plt.show()
```
