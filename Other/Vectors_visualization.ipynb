```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
plt.figure(figsize=[14, 5])
plt.style.use('seaborn-whitegrid')

indexes = [1000, 2000, 3000, 4000]
plt.xticks(ticks=indexes, labels=indexes, fontsize='large')
plt.yticks(ticks=indexes, labels=indexes, fontsize='large')
plt.xlabel('data', fontsize='large')
plt.ylabel('computer', fontsize='large')

plt.xlim(0, 4500)
plt.ylim(0, 4500)

X=[0]
Y=[0]

plt.quiver(X, Y, 1683, 1670, color='g', scale=1, scale_units='xy', angles='xy', label='digital')
plt.quiver(X, Y, 3982, 3325, color='b', scale=1, scale_units='xy', angles='xy', label='information')

plt.annotate('digital', xy=[1000, 1700], horizontalalignment='center', fontstyle='italic',
             fontweight='demi', fontsize='large', color='g')
plt.annotate('[1683,1670]', xy=[1000, 1400], horizontalalignment='center', fontstyle='normal',
             fontweight='demi', fontsize='large', color='g')

plt.annotate('information', xy=[3000, 3200], horizontalalignment='center', fontstyle='italic',
             fontweight='demi', fontsize='large', color='b')
plt.annotate('[3982,3325]', xy=[3000, 2900], horizontalalignment='center', fontstyle='normal',
             fontweight='demi', fontsize='large', color='b')

plt.savefig('Plots/example_vectors.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=[14, 5])
plt.style.use('seaborn-whitegrid')

x_indexes = [-500, 0, 500, 1000, 1500, 2000, 2500, 3000]
y_indexes = [-100, 0, 100, 200, 300, 400, 500]
plt.xticks(ticks=x_indexes, labels=x_indexes, fontsize='large')
plt.yticks(ticks=y_indexes, labels=y_indexes, fontsize='large')
plt.xlabel('computer', fontsize='large')
plt.ylabel('pie', fontsize='large')

plt.xlim(-100, 3500)
plt.ylim(-40, 500)

X=[0]
Y=[0]

plt.quiver(X, Y, 3325, 5, color='b', scale=1, scale_units='xy', angles='xy', label='information')
plt.quiver(X, Y, 1670, 5, color='g', scale=1, scale_units='xy', angles='xy', label='digital')
plt.quiver(X, Y, 2, 442, color='r', scale=1, scale_units='xy', angles='xy', label='cherry')

plt.annotate('digital', xy=[1000, 70], horizontalalignment='center', fontstyle='italic',
             fontweight='demi', fontsize='large', color='g')
plt.annotate('[1670,5]', xy=[1000, 30], horizontalalignment='center', fontstyle='normal',
             fontweight='demi', fontsize='large', color='g')

plt.annotate('information', xy=[2500, 70], horizontalalignment='center', fontstyle='italic',
             fontweight='demi', fontsize='large', color='b')
plt.annotate('[3325,5]', xy=[2500, 30], horizontalalignment='center', fontstyle='normal',
             fontweight='demi', fontsize='large', color='b')

plt.annotate('cherry', xy=[200, 310], horizontalalignment='center', fontstyle='italic',
             fontweight='demi', fontsize='large', color='r')
plt.annotate('[2,442]', xy=[200, 270], horizontalalignment='center', fontstyle='normal',
             fontweight='demi', fontsize='large', color='r')

plt.savefig('Plots/example_vectors2.png', bbox_inches='tight')
plt.show()
```
