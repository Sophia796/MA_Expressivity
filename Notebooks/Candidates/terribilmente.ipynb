### Hilfsfunktionen

```python
# Funktion, die die Sätze ausgibt, in der ein bestimmtes Wort vorkommt

def find_sentences_with_keyword(document, keyword):
    
    sentlist = []
    
    for sentence in document:
        
        if keyword.lower() in sentence.lower():
            sentence = re.sub('\n', ' ', sentence)          
            sentlist.append(sentence)
                         
    return sentlist
```

```python
def return_sentences_with_keyword(text):
    
    document = nltk.sent_tokenize(text)
    res = find_sentences_with_keyword(document, 'terribilmente')
    result = list(res)
    
    return result
```
